```python
#This program takes a file and runs stats on it such as the number of items and the total


#Defined variables
running_count = 0
running_sum = 0

print("Automated cash register")
# Get input file and open it for reading
filename = input("Enter file of prices:")
infile = open(filename) #open file

# Loop and compute the data values
for line in infile:
    line = line.strip()
    num = float(line)
    running_count = running_count + 1
    running_sum = running_sum + num

#print totals
print("File contained", str(running_count), "items totaling " + str("${:.2f}".format(running_sum)))
```

```python
#This program takes a file and runs stats on it such as the number of items and the total skipping any lines and removing dollar signs


#Defined variables
running_count = 0
running_sum = 0

print("Automated cash register")
# Get input file and open it for reading
filename = input("Enter file of prices:")
infile = open(filename) #open file

# Loop and compute the data values
for line in infile:
    try: #allows to skip over lines of input that cannot be converted to floating point numbers
        line = line.strip("$") #strips $ from line
        num = float(line) # converts to float so total can be added to
        running_count = running_count + 1
        running_sum = running_sum + num
    except: #if code under try does not run the program skips it and continues
        continue

#print totals
print("File contained", str(running_count), "items totaling " + str("${:.2f}".format(running_sum)))
```

```python

#This program validates a number as valid if it is in the correct format


phone = input("Enter number as ### ###-####:")        
valid = len(phone) == 12 #set lenght
pos = 0
while valid and pos < 12: #make a loop
    if pos == 3: #at 3rd digit there is a space this validates that
        valid = phone[pos] == " "
    elif pos == 7:
        valid = phone[pos] == "-" #at 7th digit there is a - this validates that
    else:
        valid = phone[pos].isdigit() #at every other position for it to be valid there has to be a digit
    pos = pos + 1

if valid:
    print("Valid")
else:
    print("Invalid")
```
