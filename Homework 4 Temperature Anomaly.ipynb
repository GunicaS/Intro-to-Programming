```python
#This program grabs the temperatures from the user inputted file and using the window size and prints the average temperature for that window 

filename = input("Temperature anomaly filename:") #get the input for the filename
infile = open (filename) #open the file 
infile.readline() #ignore the header line for the file

nums = [] #define the empty list

#set up a for loop 
for line in infile:
    line = line.strip() #strips the whitespace
    year,temp = line.split(",")#Since the line contains multiple values we split to get values as seperate strings of year and temp
    #print ((year),float(temp)) 
    num = float(temp) #converts the temp to a float
    nums.append(num) #adds the number to the end of the list since its a loop it keeps going till it adds all the temps to the list
#print(nums) #prints the list out

#copied code from Canvas
k = int(input("Enter window size:")) #converted to integer so the next operations can be done
index = k
year = 1880 + index
avg = sum(nums[index-k:index+k+1]) / (2*k+1)

print(year,"{:.4f}".format(avg),sep=",") #printed year and the average (which has been formatted to 4 decimal points)
```

```python

#This program gets a filename from the user and prints out the minimum and maximum in this case this is for temperature

filename = input("Temperature anomaly filename:") #get the input for the filename
infile = open (filename) #open the file 
infile.readline() #ignore the header line for the file

#Define the statistics that will be used
min_temp = None
min_year = None

max_temp = None
max_year = None

#set up a for loop 
for line in infile:
    line = line.strip() #strips the whitespace
    year,temp = line.split(",")#Since the line contains multiple values we split to get values as seperate strings of year and temp

    #print ((year),float(temp)) #prints the year and prints the temp after its converted to a float

    temp = float(temp) #convert the temperatures to floats from strings so that min and max can be counted
    if min_temp == None or temp < min_temp: #runs through to find the min value
        min_temp = temp
        min_year = year
    
    if max_temp == None or temp > max_temp: #runs through to find the max value
        max_temp = temp
        max_year = year
        
print ("Min temp:" ,min_temp, "in", min_year) #prints the minimum temperature and its corresponding year
    
print ("Max temp:" ,max_temp, "in", max_year) #prints the maximum temperature and its corresponding year
```
