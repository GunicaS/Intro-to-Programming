```python

#This program spell checks disregarding punctuation and capitalization
import string

def main():
    # read the list of correctly spelled words from a file
    wordList = read_word_list()

    print(len(wordList), "words read into spelling dictionary")

    # check sentence
    check_sentence(wordList)

def read_word_list():
    wordList = []
    infile = open("scrabble.txt")
    while True:
        line = infile.readline().strip()
        if not line:
            break
        wordList.append(line)
    return wordList

def check_sentence(wordList):
    sentence = input("Enter a sentence to check:")
    words = sentence.split()
    
    incorrect_words = []
    for word in words:
        word = word.lower() #word is made sure to be lowercased
        word = word.strip(string.punctuation)#puncuation is removed
        if word not in wordList:
            incorrect_words.append(word)
        
    for incorrect_word in incorrect_words:
        print(incorrect_word)


main()
```
